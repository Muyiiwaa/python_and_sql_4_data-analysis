### DATA VISUALIZATION WITH SEABORN MATPLOTLIB AND PLOTLY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# load the dataset
insurance_df = pd.read_csv('insurance.csv')
insurance_df.head()
```

### BAR CHART

```python
sns.set_style('darkgrid')
sns.set_palette('bright')
```

```python
# SEABORN
plt.figure(figsize=(20,20))
sns.barplot(data=insurance_df, x = 'sex', y = 'bmi', hue = 'sex')
plt.title(label='Average BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('Average BMI')
```

```python
# MATPLOTLIB

temp_df = insurance_df[['sex','bmi']].groupby(by='sex').mean()
temp_df

plt.bar(x = temp_df.index, height=temp_df['bmi'], color = 'red')
plt.title(label='Average BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('Average BMI')
plt.figure(figsize=(5,5))
```

```python
## PLOTLY

px.histogram(data_frame=insurance_df, x = 'sex', y = 'bmi',
             histfunc='avg', title='Average BMI by Sex',
             color= 'sex',height=600, width=800)
```

### SCATTER PLOT

```python
insurance_df.head()
```

```python
# seaborn

sns.scatterplot(data=insurance_df, x = 'bmi', y = 'charges',
                hue= 'smoker')
```

```python
# plotly

px.scatter(data_frame=insurance_df, x = 'bmi', y = 'charges',
           color = 'smoker')
```

## PIE CHART

```python
plt.pie(x = temp_df['bmi'], labels=temp_df.index, autopct='%.2f%%')
plt.title('Pie Chart of BMI vs SEX')
```

```python
# plotly

px.pie(data_frame=insurance_df, names='sex', values='bmi',
       title='pie chart of bmi vs sex')
```
